```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
iris_df = sns.load_dataset('iris')
```

```python
iris_df
```

```python
sns.scatterplot(data = iris_df , x = 'sepal_length' , y = 'petal_width',hue = 'species')
```

```python
numeric_cols = ['sepal_length' , 'sepal_width' , 'petal_length' , 'petal_width']
```

```python
x = iris_df[numeric_cols]
```

**K-MEANS CLUSTERING**

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 3 , random_state = 42)
```

```python
model.fit(x)
```

```python
model.cluster_centers_
```

```python
preds = model.predict(x)
preds
```

```python
sns.scatterplot(data = x , x = 'sepal_length' , y = 'petal_length' , hue = preds)
centre_x , centre_y = model.cluster_centers_[:,0] , model.cluster_centers_[:,2]
plt.plot(centre_x , centre_y , 'xb')
```

```python
model.inertia_
```

**DBSCAN**

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps = 1.1 , min_samples = 4)
```

```python
model.fit(x)
```

```python
model.labels_
```

```python
sns.scatterplot(data = x , x = 'sepal_length' , y = 'petal_length' , hue = model.labels_)
```

```python
from fastai.collab import CollabDataBunch , collab_learner
```

```python
cols = ['User ID' , 'Movie ID' , 'Rating' , 'Timestamp']
ratings_df = pd.read_csv(zip_file_path , delimiter = '\t' , header = None , names = cols)
ratings_df.head()
```

```python
ratings_df['User ID'] = ratings_df['User ID'].astype('category')
ratings_df['Movie ID'] = ratings_df['Movie ID'].astype('category')
ratings_df['Rating'] = ratings_df['Rating'].astype('category')
ratings_df['Timestamp'] = ratings_df['Timestamp'].astype('category')
```

```python
data = CollabDataBunch.from_df(ratings_df , valid_pct = 0.1)
data.show_batch()
```

```python
learn = collab_learner(data , n_factors = 40 , y_range=[0,5.5],wd=.1)
```

```python
learn.lr_find()
learn.recorder.plot(skip_end = 15)
```

```python
learn.fit_one_cycle(5 , 0.01)
```

```python
(users , items) , ratings = next(iter(data.valid_dl))
preds = learn.model(users , items)
for p in list(zip(users , items , ratings , preds))[:16]:
    print('{}\t{}\t{}\t{:.1f}\t{:.1f}'.format(p[0],p[1],p[2],p[3],p[3]-p[2]))
```
